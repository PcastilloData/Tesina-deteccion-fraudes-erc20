## 0-1 Eliminación de columnas irrelevantes
Variables con demasiados nulos (>90%)

name_y → 98% nulos → se elimina porque no aporta información confiable.

Variables con valores casi únicos (>90%)

first_seen, last_seen (≈99% únicos) → casi identificadores temporales, difícil generalizar, se descartan.

name_x, id, symbol se descartan es más relevante el address es más especifico.

Variables categóricas con muy pocos valores únicos

organization (2 valores) e id (49 valores) → bajo poder predictivo, se descartan.

Variables financieras (current_price, market_cap, total_volume)

Aunque parecen importantes, muestran <1% de variabilidad en toda la muestra → no diferencian entre tokens → se descartan (probable error de merge o baja cobertura).

```python
# Copia del dataframe original
df_1 = df.copy()

# Columnas a eliminar según criterio justificado
drop_cols = [
    "name_y", "name_x", "symbol", "first_seen", "last_seen",
    "organization", "id", "current_price", "market_cap", "total_volume"
]

df_1 = df_1.drop(columns=drop_cols, errors="ignore")

# Convertir posibles numéricas que vienen como 'object'
num_cols = [
    "num_transfers", "approx_holders", "avg_transfers_per_day",
    "std_transfers_per_day", "active_days", "age_days",
    "inactive_days_proxy", "top10_concentration_proxy"
]
for col in num_cols:
    if col in df_1.columns:
        df_1[col] = pd.to_numeric(df_1[col], errors="coerce")

print("Preprocesamiento listo. Shape final:", df_1.shape)
print("Columnas finales:", df_1.columns.tolist())
```

## 0-2 Revisión de variables restantes

```python
# Candidatas tras limpieza inicial
cols_revisar = [
    "address", "approx_holders", "num_transfers",
    "std_transfers_per_day", "avg_transfers_per_day",
    "active_days", "age_days", "inactive_days_proxy",
    "top10_concentration_proxy", "t2-labels", "t1-labels"
]

df_check = df_1.copy()

#eliminar etiquetas faltantes
df_check = df_check[~((df_check["t1-labels"].astype(str) == "0") | (df_check["t2-labels"].astype(str) == "0"))]

# Intentar conversión de object a numérico (sin dañar categóricas claras)
num_like = [
    "approx_holders", "num_transfers", "std_transfers_per_day",
    "avg_transfers_per_day", "active_days", "age_days",
    "inactive_days_proxy", "top10_concentration_proxy"
]
for col in num_like:
    df_check[col] = pd.to_numeric(df_check[col], errors="coerce")

# Resumen detallado
def resumen_columnas(df):
    """
    Muestra un resumen de cada columna del DataFrame: tipo, % de NaN, cantidad de valores únicos
    y top 5 valores más frecuentes.
    """
    for col in df.columns:
        print(f"\n--- {col} ---")
        print(f"Tipo: {df[col].dtype}")
        print(f"% NaN: {df[col].isna().mean() * 100:.2f}")
        print(f"Valores únicos: {df[col].nunique()}")

        top_vals = df[col].value_counts(dropna=False).head(5)
        print("Top 5 valores más frecuentes:")
        print(top_vals.to_dict())
resumen_columnas(df_check)
```

## 0-3 0-4 Conversión de tipos y Tratamiento de valores faltantes

1.   Elemento de lista
2.   Elemento de lista



```python
df_3 = df_check.copy()


# Conversión de tipos
# -----------------------------
num_cols = ['num_transfers','approx_holders','top10_concentration_proxy',
            'avg_transfers_per_day','std_transfers_per_day',
            'active_days','age_days','inactive_days_proxy']

for col in num_cols:
    df_3[col] = pd.to_numeric(df_3[col], errors='coerce')  # fuerza numérico

cat_cols = ['address','t1-labels','t2-labels']
for col in cat_cols:
    df_3[col] = df_3[col].astype('category')

# -----------------------------
# 4Tratamiento de valores faltantes y valores inválidos
# -----------------------------

# std_transfers_per_day: NaN → 0
df_3['std_transfers_per_day'] = df_3['std_transfers_per_day'].fillna(0)

# approx_holders: si hay 0 → imputar con mediana
df_3['approx_holders'] = df_3['approx_holders'].replace(0, np.nan)
df_3['approx_holders'] = df_3['approx_holders'].fillna(df_3['approx_holders'].median())

# inactive_days_proxy: valores negativos → NaN, imputar con mediana
df_3['inactive_days_proxy'] = df_3['inactive_days_proxy'].apply(lambda x: x if x >= 0 else np.nan)
df_3['inactive_days_proxy'] = df_3['inactive_days_proxy'].fillna(df_3['inactive_days_proxy'].median())

# active_days: si aparece 0 → considerar NaN e imputar con mediana
df_3['active_days'] = df_3['active_days'].replace(0, np.nan)
df_3['active_days'] = df_3['active_days'].fillna(df_3['active_days'].median())

# top10_concentration_proxy: valores negativos → NaN e imputar con mediana
df_3['top10_concentration_proxy'] = df_3['top10_concentration_proxy'].apply(lambda x: x if x >= 0 else np.nan)
df_3['top10_concentration_proxy'] = df_3['top10_concentration_proxy'].fillna(df_3['top10_concentration_proxy'].median())


print(df_3.info())
print(df_3.head())


```

Primero quitamos columnas con >90% nulos, porque casi no aportan y

---

entorpecen imputación.

---



La conversión a numérico se hace solo si una alta fracción de valores se convierten correctamente (evita convertir symbol o address por error).

Las categóricas de baja cardinalidad se transforman a category (útil para codificación posterior y memoria).

Imputación: mediana para numéricas (robusta ante outliers), modo o "unknown" para categóricas según cardinalidad.

```python

df_5 = df_3.copy()

le_t1 = LabelEncoder()
df_5['t1-labels_encoded'] = le_t1.fit_transform(df_5['t1-labels'])

le_t2 = LabelEncoder()
df_5['t2-labels_encoded'] = le_t2.fit_transform(df_5['t2-labels'])
```

## 0-6 Detección y tratamiento de outliers

```python
df_6 = df_5.copy()

num_cols = df_6.select_dtypes(include='number').columns.tolist()
# Excluir columnas de labels codificadas
num_cols = [col for col in num_cols if col not in ['t1-labels_encoded', 't2-labels_encoded']]
print(df_6[num_cols].describe())
```

```python

# Tratar outliers usando percentiles 2 y 98
for col in num_cols:
    lower = df_6[col].quantile(0.02)    #se achico más de lo normal debido a gran cantidad de outlier
    upper = df_6[col].quantile(0.98)
    df_6[col + '_capped'] = df_6[col].clip(lower, upper)

# Aplicar log-transform a las variables sesgadas
log_cols = [
    'num_transfers_capped',
    'approx_holders_capped',
    'top10_concentration_proxy_capped',
    'avg_transfers_per_day_capped',
    'std_transfers_per_day_capped',
    'active_days_capped',
    'age_days_capped',
    'inactive_days_proxy_capped'
]

for col in log_cols:
    df_6[col + '_log'] = np.log1p(df_6[col])

# Resumen estadístico post-tratamiento
summary = df_6[[c + '_log' for c in log_cols]].describe().T
summary['skew'] = df_6[[c + '_log' for c in log_cols]].skew()
summary['kurtosis'] = df_6[[c + '_log' for c in log_cols]].kurtosis()

print(summary)
```
